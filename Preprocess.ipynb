```python
datalist=[]
with open("WikipediaTrain_after","r") as f:
    for l in f.readlines():
        tmp = l[:-2].split("|")
        assert len(tmp)==4
        datalist.append(tmp)
print(datalist[420][0],datalist[420][1])
```

```python
import json
import torch
import torch.nn.functional as F
from fairseq.models.roberta import RobertaModel
roberta = torch.hub.load('pytorch/fairseq', 'roberta.large')
roberta.eval()  # disable dropout
roberta.cuda()  # use the GPU (optional)

examples=[]
for data in datalist:
    doc = roberta.extract_features_aligned_to_words(data[1])
    for token in doc:
        if str(token)==data[2]:
            examples.append((token.vector,data[3]))
            break
print(examples)
```
